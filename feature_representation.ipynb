```python
import h5py
import pandas as pd
import numpy as np
import gc
```

## Read the extracted features

```python
# Baseline Path
# filename = "baseline_cityscape"
# filename = "baseline_itri"

# Adversarial Path
# filename = "adversarial_cityscape"
# filename = "adversarial_itri"

# Finetune Path
# filename = "finetune_cityscape"
# filename = "finetune_itri"

# Oracle Path
# filename = "oracle_cityscape"
filename = "oracle_itri"

path = "extracted_features/" + filename + ".hdf5"

print(path)
```

```python
# Read the hdf5 file
with h5py.File(path, "r") as f:
    # List all groups
    print("Keys: %s" % f.keys())
    a_group_key = list(f.keys())[0]
    
    # Get the data
    data = list(f[a_group_key])
    
# Convert the data into dataframe
df = pd.DataFrame(np.array(data))

# Remove data to save memory
del data
gc.collect()
```

```python
df
```

```python
# Print the class distinct value
classValue = df[4096].unique()
print(classValue)
```

```python
# Remove the background class > last column == 0
df_cleaned = df[df[4096] != 0]
df_cleaned
```

```python
# Write into smaller CSV
out_path = "extracted_features/csv/" + filename + ".csv"
df_cleaned.to_csv(out_path, index=False, encoding='utf-8')
```

```python
# Remove df to save memory
del df
gc.collect()
```

```python
x_features = df_cleaned.iloc[: , :-1]
y_labels = df_cleaned[4096]
```

## Dimensionality Reduction

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.manifold import TSNE
import umap
import umap.plot
```

```python
# Principal Component Analysis
pca = PCA(n_components=2)
x_features_pca = pca.fit(x_features).transform(x_features)
```

```python
# Percentage of variance explained for each components
print('Explained variance ratio (first two components): %s' % str(pca.explained_variance_ratio_))
x_features_pca
```

```python
# Linear Discriminant Analysis
lda = LinearDiscriminantAnalysis(n_components=2)
x_features_lda = lda.fit(x_features, y_labels).transform(x_features)
```

```python
x_features_lda
```

```python
# T-SNE
x_features_tsne = TSNE(learning_rate=100).fit_transform(x_features)
```

```python
x_features_tsne
```

```python
# UMAP
umap_mapper = umap.UMAP().fit(x_features)
```

```python
umap.plot.points(umap_mapper, labels=y_labels)
# umap.plot.points(umap_mapper, labels=y_labels, color_key_cmap='Paired', background='white')
# umap.plot.diagnostic(umap_mapper, diagnostic_type='pca')
```

```python
# Plot the result
target_labels = np.unique(y_labels)
colors = ['navy', 'turquoise', 'darkorange', 'red']
lw = 2
alpha = .4

# PCA
plt.figure()
for color, i, target_label in zip(colors, target_labels, target_labels):
    plt.scatter(x_features_pca[y_labels == i, 0],
                x_features_pca[y_labels == i, 1],
                color=color,
                lw=lw,
                label=target_label,
                alpha=.1
                )
    plt.legend(loc='best', shadow=False, scatterpoints=1)
    plt.title('PCA of Cityscape (Baseline) Dataset')

# LDA
plt.figure()
for color, i, target_label in zip(colors, target_labels, target_labels):
    plt.scatter(x_features_lda[y_labels == i, 0],
                x_features_lda[y_labels == i, 1],
                color=color,
                lw=lw,
                label=target_label,
                alpha=.1
                )
    plt.legend(loc='best', shadow=False, scatterpoints=1)
    plt.title('LDA of Cityscape (Baseline) Dataset')

# TSNE
plt.figure()
for color, i, target_label in zip(colors, target_labels, target_labels):
    plt.scatter(x_features_tsne[y_labels == i, 0],
                x_features_tsne[y_labels == i, 1],
                color=color,
                lw=lw,
                label=target_label,
                alpha=.1
                )
    plt.legend(loc='best', shadow=False, scatterpoints=1)
    plt.title('TSNE of Cityscape (Baseline) Dataset')

plt.show()
```
